```python
import os
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Define the path to your dataset
dataset_path = 'C:/Users/Neha R/OneDrive/Desktop/DentalXrays/DentalXrays/preview'

# Function to load and display images
def load_and_display_images(dataset_path):
    # Get a list of all files in the directory
    files = os.listdir(dataset_path)

    # Display the first 20 images (original and thresholded)
    plt.figure(figsize=(30, 10))
    for i in range(20):
        img_path = os.path.join(dataset_path, files[i])
        img = mpimg.imread(img_path)

        # Display original image
        plt.subplot(4, 10, i + 1)
        plt.imshow(img)
        plt.axis('off')
        plt.title(f'Original Image {i + 1}')

        # Convert the image to grayscale
        gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        # Apply a simple thresholding (you can adjust the threshold value)
        _, threshold_img = cv2.threshold(gray_img, 128, 255, cv2.THRESH_BINARY)

        # Display thresholded image
        plt.subplot(4, 10, i + 21)
        plt.imshow(threshold_img, cmap='gray')
        plt.axis('off')
        plt.title(f'Threshold Image {i + 1}')

    plt.show()

# Call the function
load_and_display_images(dataset_path)
```
